```python
global deg, weights
    
#Shuffling
df = pd.read_csv("fods_2.csv").sample(frac = 1, random_state = 99)
#Standardization - Z Score
df.iloc[ : , 0:-1] = df.iloc[ : , 0:-1].apply(lambda x: (x - x.mean()) / x.std(), axis = 0)
#80-20 Train Test
train_len = int(0.8*len(df))
train, test = np.split(df.to_numpy(), [train_len])
feature_count = df.shape[1] - 1
  
X_train = np.delete(train, -1, 1)
Y_train = train[:, -1]
    
X_test = np.delete(test, -1, 1)
Y_test = test[:, -1]
```

```python
features_chosen = list(range(0, feature_count))
for x in range(feature_count + 1):
    error = gdc.linear_regression(X_train[: , features_chosen], Y_train, X_test[: , features_chosen], Y_test)
    new_error = 1e9
    new_feature = -1
    for i in features_chosen:
        features_chosen_copy = features_chosen.copy()
        print(features_chosen_copy)
        features_chosen_copy.remove(i)
        curr_error = gdc.linear_regression(X_train[: , features_chosen_copy], Y_train, X_test[: , features_chosen_copy], Y_test)
        if curr_error < new_error:
            new_error = curr_error
            new_feature = i
    if new_error < error:
        features_chosen.remove(new_feature)
    else:
        break
    
print(len(features_chosen))
print(features_chosen)
```
